# Impulse Response Analysis

```python
from statsmodels.tsa.api import VAR
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import statsmodels
statsmodels.__version__
```

```python
df = pd.read_csv('datasets/garch_volatility_df.csv', index_col=0)
df.index = pd.to_datetime(df.index)
df.columns
```

```python
data = df[['Geopolitics', 'New Energy', 'Green Bond',
       'Carbon Price', 'Crude Oil']]
model = VAR(data)
```

```python
results = model.fit(2)
```

```python
results.plot()
plt.show
```

```python
irf = results.irf(24)
```

```python
irf.plot(orth=False,
         impulse='Geopolitics')
plt.show()
```
